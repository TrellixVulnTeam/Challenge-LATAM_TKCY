```python
import shap
import random
import pandas as pd
import numpy as np
import itertools
import matplotlib.pyplot as plt

import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

shap.initjs()

random.seed(4242)
np.random.seed(4242)
```

```python
def plot_roc_curve(y, scores):
    auc = round(100 * roc_auc_score(y, scores))
    fpr_rt, tpr_rt, _ = roc_curve(y, scores)
    plt.plot(fpr_rt, tpr_rt, label='RT + LR')
    plt.title(f"AUC {auc}%")
    plt.show()
    
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
class CaterogicalEncoder:
    
    """
    Este encoder reemplaza el valor de una categoria por el promedio del target para esa caterogia
    """
    
    def __init__(self, cols, missing_key="general"):
        self.cols = cols
        self.mapping = {}
        self.missing_key = missing_key
        
    def map_values(self, col, x):
        if x in self.mapping[col]:
            return self.mapping[col][x]
        else:
            if self.missing_key == "general":
                return self.mapping[col]["general"]
    
    def fit(self, X, y):
        for col in self.cols:
            self.mapping[col] = {}
            for value in X[col].unique():
                mean = y[X[col] == value].mean()
                self.mapping[col][value] = mean
            self.mapping[col]["general"] = y.mean()
                
    def transform(self, X, y=None):
        X = X.copy()
        for col in self.cols:
            X[col] = X[col].apply(lambda x: self.map_values(col, x))
        return X
```

```python
df = pd.read_csv("data.csv", sep="|")
display(df.head())
```

# P4. . Entrena uno o varios modelos (usando el/los algoritmo(s) que prefieras) para estimar la probabilidad de atraso de un vuelo

```python
X = df.drop(labels="atraso_15", axis=1)
y = df["atraso_15"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

### Baseline

Todos 1

```python
y_test_pred = np.ones(y_test.shape[0])
y_test_scores = np.ones(y_test.shape[0])

cnf_matrix = confusion_matrix(y_test, y_test_pred)

pre = round(  100 * precision_score( y_test, y_test_pred) , 2)
rec = round(  100 * recall_score( y_test, y_test_pred) , 2)
plot_roc_curve(y_test, y_test_scores)

print(f"Precision: {pre}%")
print(f"Recall: {rec}%")
plot_confusion_matrix(cnf_matrix, classes=["0_", "1_"],
                      title='Confusion matrix, without normalization')
plt.show()
```

### Modelo 1 - Custom Encoder

```python
CATEGORICAL_COLS = ["OPERA", "SIGLADES", "TIPOVUELO", "estacion-I", "periodo_dia"]
```

```python
cat_encoder = CaterogicalEncoder(CATEGORICAL_COLS)
cat_encoder.fit(X_train, y_train)
x_train = cat_encoder.transform(X_train)
x_test = cat_encoder.transform(X_test)
```

```python
model = xgb.XGBClassifier(
             learning_rate =0.1,
             n_estimators=100,
             max_depth=10,
             min_child_weight=1,
             gamma=0,
             subsample=0.8,
             colsample_bytree=0.2,
             objective= 'binary:logistic',
             nthread=4,
             scale_pos_weight=1,
             seed=27)

model.fit(x_train, y_train)
y_test_pred = model.predict(x_test)
y_test_scores = model.predict_proba(x_test)
y_test_scores = [x[1] for x in y_test_scores]
cnf_matrix = confusion_matrix(y_test, y_test_pred)


pre = round(  100 * precision_score( y_test, y_test_pred) , 2)
rec = round(  100 * recall_score( y_test, y_test_pred) , 2)
plot_roc_curve(y_test, y_test_scores)

print(f"Precision: {pre}%")
print(f"Recall: {rec}%")
plot_confusion_matrix(cnf_matrix, classes=["0_", "1_"],
                      title='Confusion matrix, without normalization')
plt.show()
```

```python
xgb.plot_importance(model)
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(x_train)
# summarize the effects of all the features
shap.summary_plot(shap_values, x_train)
```

### Modelo 2 - One Hot

```python
CATEGORICAL_COLS = ["OPERA", "SIGLADES", "TIPOVUELO", "estacion-I", "periodo_dia"]
```

```python
cat_encoder = OneHotEncoder(handle_unknown='ignore')


X_train_categorical = X_train[CATEGORICAL_COLS]
X_test_categorical = X_test[CATEGORICAL_COLS]

cat_encoder.fit(X_train_categorical)

X_train_categorical = cat_encoder.transform(X_train_categorical).toarray()
X_test_categorical = cat_encoder.transform(X_test_categorical).toarray()



x_train = pd.concat([X_train.drop(labels=CATEGORICAL_COLS, axis=1).reset_index(drop=True), 
                                   pd.DataFrame(X_train_categorical,columns=cat_encoder.get_feature_names())], axis=1)

x_test = pd.concat([X_test.drop(labels=CATEGORICAL_COLS, axis=1).reset_index(drop=True), 
                                   pd.DataFrame(X_test_categorical,columns=cat_encoder.get_feature_names())], axis=1)

```

```python
model = xgb.XGBClassifier(
             learning_rate =0.1,
             n_estimators=100,
             max_depth=10,
             min_child_weight=1,
             gamma=0,
             subsample=0.8,
             colsample_bytree=0.2,
             objective= 'binary:logistic',
             nthread=4,
             scale_pos_weight=1,
             seed=27)

model.fit(x_train, y_train)
y_test_pred = model.predict(x_test)
y_test_scores = model.predict_proba(x_test)
y_test_scores = [x[1] for x in y_test_scores]
cnf_matrix = confusion_matrix(y_test, y_test_pred)


pre = round(  100 * precision_score( y_test, y_test_pred) , 2)
rec = round(  100 * recall_score( y_test, y_test_pred) , 2)
plot_roc_curve(y_test, y_test_scores)

print(f"Precision: {pre}%")
print(f"Recall: {rec}%")
plot_confusion_matrix(cnf_matrix, classes=["0_", "1_"],
                      title='Confusion matrix, without normalization')
plt.show()
```

```python
xgb.plot_importance(model,  max_num_features=10)
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(x_train)
# summarize the effects of all the features
shap.summary_plot(shap_values, x_train)
```

Para resolver este desafío preferí utilizar xgboost por diferentes motivos:
* Es conocido por tener buen performance

Respecto a las métricas calculé diferentes: precisión, recall, auc. Las razones son las siguientes: La curva ROC indica la evolución de la tasa de verdaderos positivos y tasa de falsos positivos para diferentes umbrales de decisión. Esto es útil cuando se quiere establecer un modo de operación que privilegie el recall o precisión.
 
La precisión indica que tan confiable es la clasificación del modelo, mientras que el recall indica que tan sensible es el modelo. En general, se busca tener un buen equilibro de estos mediante el F1-Score pero en está ocasión preferí mantenerlos por separado para tener mayor profundidad de análisis.
 
En esta ocasión entrené dos modelos XGBOOST, donde la principal diferencia está en la forma de encodear las variables categóricas. El primer modelo hace uso de un método que reemplaza cada valor de una variable categórica por el promedio de la variable objetivo. La idea de esto es no asignar números de forma aleatoria como ocurre con un encoding tradicional, sino que la idea es mantener el orden según la relación que tenga la variable con el target. El segundo modelo utiliza one hot encoding para cada variable categórica (una columna por cada valor).
Adicionalmente, deje un baseline (todas las predicciones son 1) para tener un punto de referencia (AUC 50%, precisión 44,29%, recall 100%).
 
 
El primer modelo alcanzó un AUC de 64%, una precisión de 60.3% y un recall de 40.2%. Estas medidas indican que el 60% de las detecciones de atrasos en los vuelos del modelo son reales. Por otra parte, el modelo está capturando el 40% de todos los atrasos.
 
El segundo modelo alcanzó un AUC de 64%, una precisión de 60.27% y un recall de 39.7%. Al comparar este modelo con el primero se ve que sus detecciones son casi igual de confiables que el primero, pero es levemente (0.5%) menos sensible que el primero. Si utilizaramos el F1 Score para decidir qué modelo usar, el primero sería el ganador.
 
Para obtener la importancia de las variables usé dos métodos, el F-score de XGBOOST y los SHAP values. En el primer modelo, las variables más importantes son el destino, el día, la hora y el mes, mientras que en el segundo son el día de la semana, día, hora y mes. Por otro lado, en el gráfico de los shap values podemos ver cómo contribuye cada variable en la probabilidad de atraso. Por ejemplo en el primer modelo, valores altos de OPERA, TIPOVUELO, periodo_dia, SIGLADES aumentan la probabilidad de atraso. Esto tiene mucho sentido debido al tipo de encoding que se está aplicando. En el segundo modelo en cambio, al tener variables dummies se pueden ver los efectos de cada destino/aerolínea/etc por separado. Por ejemplo, si el destino es punta arenas (high value), entonces la probabilidad de retraso disminuye, si la operadora del vuelo es Aerolíneas Argentinas aumenta la probabilidad de atraso.
 
Por último, para mejorar el performance de cualquier modelo entrenado es necesario incluir más atributos como el clima, tipo de avión, cantidad de pasajeros, si el avión viene de otro vuelo, la saturación del aeropuerto y probablemente un montón de otros atributos que puedan ser mucho mejores indicadores que el dia de la semana o la estación del año o el mes. Desde el punto de vista del modelo, también se debe hacer un finetuning de los modelos, también se debe probar con otros modelos, y antes de pasar a producción estos deben ser validados en diferentes ventanas de tiempo para evitar que la estacionalidad de los datos perjudique el desempeño del modelo (por ejemplo, entrenar con 3 semanas consecutivas y validar con la siguiente, de esta forma el modelo debe mantenerse entrenado siempre con 3 semanas de datos).
